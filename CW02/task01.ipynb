```python
class Solution:
    @staticmethod
    def maxProfit(prices: list[int]) -> int:
        max = 0
        l = 0
        r = 1
        while r < len(prices):
            if prices[r] < prices[l]:
                l = r
            else:
                profit = prices[r] - prices[l]
                if profit >= max:
                    max = profit
            r+=1
        return max
```

```python
print(Solution.maxProfit([1,2,3,4,5,6,7]))
```
